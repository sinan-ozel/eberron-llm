```python
import os
import json
from glob import glob

from mangoCR import pdf2image_ocr
from pypdf import PdfReader
from pypdf.errors import PdfStreamError

from tqdm.notebook import tqdm
```

```python
folder = '/corpus/pdf'
corpus_folder = '/corpus/texts_v4'
```

```python
for absolute_path in tqdm(glob(f'{folder}/**/*.pdf', recursive=True)):
    path, filename = os.path.split(absolute_path)
    _, subfolder = os.path.split(path)
    output_file_path = os.path.join(corpus_folder, f'{filename[:-4]}.md')
    if not (os.path.exists(output_file_path) and os.path.getsize(output_file_path) > 0):
        pdf2image_ocr(absolute_path, output_file=output_file_path)
```

```python
edition = {}
content_category = {}
pdf_metadata = {}
for absolute_path in tqdm(glob(f'{folder}/**/*.pdf', recursive=True)):
    path, filename = os.path.split(absolute_path)
    reader = PdfReader(absolute_path)
    pdf_metadata[filename] = reader.metadata if reader.metadata is not None else {}
    subfolder = path.split('/')[-1]
    content_category[filename] = subfolder
    edition[filename] = ''
    for ed in ['3e', '4e', '5e']:
        if ed in subfolder:
            edition[filename] = ed
            break
```

```python
metadata = []
for file_name in pdf_metadata:
    if pdf_metadata[file_name] is None:
        print(file_name)
    metadata.append(
        {
            'filename': file_name,
            'edition': edition[file_name],
            'book_category': content_category[file_name],
            'pdf/title': str(pdf_metadata[file_name].get('/Title', '')),
            'pdf/author': str(pdf_metadata[file_name].get('/Author', '')),
        }
    )
```

```python
with open(os.path.join(corpus_folder, 'metadata.json'), 'w') as metadata_file:
    json.dump(metadata, metadata_file)
```
